**The code below reads the link and prints out the information in it**

```python
data = pd.read_csv('https://tools.wprdc.org/downstream/76fda9d0-69be-4dd5-8108-0de7907fc5a4')
print (data)
```

**The code below prints the entire "Y" column of the Data Frame, then prints the first five things of the "Y" column.**

```python
print(data["Y"])
print(data["Y"].head())
```

**The code below prints out the first 5 information in the link**

```python
data.head()
```

**This code drops all the columns without a neighborhood**

```python
data[data['NEIGHBORHOOD'].notna()]
```

**This code filters out the data so that only the noise complaints show**

```python
noise_data = data[data['REQUEST_TYPE'].str.contains('Excessive Noise/Disturbance', case=False, na=False)]
print (noise_data)
```

**This code drops rows that have missing values**

```python
noise_data[data['NEIGHBORHOOD'].notna()]
```

**This code prints number of noise complaints in each neighborhood**

```python
neighborhood_counts= noise_data['NEIGHBORHOOD'].value_counts()
print(neighborhood_counts)
```

**This code groups the data by the "NEIGHBORHOOD" column and then countd the number of occurrences for each neighborhood. After that, it plots a bar chart using the counts.**

```python
realData= noise_data.groupby("NEIGHBORHOOD").count()
realData["REQUEST_ID"].plot.bar(figsize=(20,20), title=" Neighborhood Noise Complaints")
```

**The code below arranges the noise complaints in ascending order**

```python
sortedneighborhood=realData.sort_values("REQUEST_ID")
sortedneighborhood["REQUEST_ID"].plot.bar(figsize=(20,20), title=" Neighborhood Noise Complaints")
```

Why do we care about noise complaints? Well, everyone knows that 'early to bed, early to rise' makes a person healthy!
