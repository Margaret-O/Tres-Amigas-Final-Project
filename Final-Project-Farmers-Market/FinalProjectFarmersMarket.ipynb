**Our Project: Finding out what neighborhood in Pittsburgh is best to raise a D1 Basketball Player**

```python
import pandas as pd
import fpsnippets
from fpsnippets import geo_to_neighborhood
%matplotlib inline
```

![farmers](https://media.giphy.com/media/3ov9jDwb4159zNHlDy/giphy.gif)

```python
data = pd.read_csv('farmersMarket.tsv', sep= '\t' )
print(data)
```

```python
print(data["latitude"])
print(data["latitude"].head())
```

```python
print(data["longitude"])
print(data["longitude"].head())
```

```python
for index, row in data.iterrows():
    # Convert latitude and longitude to strings
    latitude_str = str(row['latitude'])
    longitude_str = str(row['longitude'])
    latitude = float(latitude_str)
    longitude = float(longitude_str)
    # Print the converted values
    print(f"Latitude: {latitude_str}, Longitude: {longitude_str}")
    print(f"Number of rows: {len(data)}")
    # Assuming fpsnippets is a method that converts latitude and longitude to neighborhood names
    # Replace this with your actual method
    neighborhood_name = geo_to_neighborhood(latitude, longitude)

    # Print the neighborhood name
    print(f"Neighborhood: {neighborhood_name}\n")
    data.loc[index, "Neighborhood"] = neighborhood_name
```
